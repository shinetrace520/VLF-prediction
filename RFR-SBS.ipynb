```python
evaluate_r_2=[]
evaluate_r_2_tr=[]

n=-1

for train_index, test_index in rs_2.split(F_0):
    scaler_M.fit(F_0[train_index,:])
    
    F_tr=scaler_M.transform(F_0[train_index,:])
    F_te=scaler_M.transform(F_0[test_index,:])
        
    n=n+1
    
    RFR=RandomForestRegressor(n_estimators=50,random_state=0,
                              max_depth=_max_depth[n],
                              min_samples_split=_min_samples_split[n])
    
    RFR.fit(F_tr,(p.reshape(-1,1)[train_index,:]-1)*100)
    
    p_tr=RFR.predict(F_tr)
    p_tr=p_tr.reshape(-1,1)
    r_2_tr=r2_score((p.reshape(-1,1)[train_index,:]-1)*100, p_tr)
    evaluate_r_2_tr.append(r_2_tr)
    
    p_pre=RFR.predict(F_te)
    p_pre=p_pre.reshape(-1,1) 
    r_2=r2_score((p.reshape(-1,1)[test_index,:]-1)*100, p_pre)
    evaluate_r_2.append(r_2)
```
